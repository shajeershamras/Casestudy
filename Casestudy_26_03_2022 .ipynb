# Top 5 car based on their sales price

```python
#To find the largest 5 cars based on sales price from database
sdb1=mdb.nlargest(5,['Sale Price'])

#To highlight the car model,make and sales price columns
def highlight_cols(x):
      
    # copy df to new - original data is not changed
    sdb1 = x.copy()
    # select all values to neutral color
    sdb1.loc[:, :] = 'background-color: '
    # overwrite values to yellow color
    sdb1[['Sale Price', 'Make', 'Model']] = 'background-color: yellow'
    # return color df
    return sdb1 
  
print("Top 5 cars based on their sell price :")
display(sdb1.style.apply(highlight_cols, axis = None))

```

# Least 5 cars based on their resell price

```python
#To find the smallest 5 cars based on resell price from database
sdb2=mdb.nsmallest(5,['Resell Price'])

#To hughlight the car model
def highlight_cols(x):
      
    # copy df to new - original data is not changed
    sdb2 = x.copy()
      
    # select all values to green color
    sdb2.loc[:, :] = 'background-color: '
      
    # overwrite values grey color
    sdb2[['Resell Price', 'Make', 'Model']] = 'background-color: yellow'
      
    # return color df
    return sdb2 
  
print("Least 5 cars based on their resell price :")
display(sdb2.style.apply(highlight_cols, axis = None))
```
